2. Add Two Numbers

Medium

You are given two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order and each of their nodes contain a single digit. Add the two numbers and return it as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.

Example:

Input: (2 -> 4 -> 3) + (5 -> 6 -> 4)
Output: 7 -> 0 -> 8
Explanation: 342 + 465 = 807.

```python
'''
transform ListNode to List v1
wtc : O()
wsc : O()
'''
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def addTwoNumbers(l1: ListNode, l2: ListNode) -> ListNode:
    def reverseNode(head: ListNode) -> ListNode:
        node, rev = head, None
        while node:
            next, node.next = node.next, rev
            rev, node = node, next
        return rev

    def nodeToStr(l1: ListNode) -> str:
        str_val = ''
        while l1:
            str_val += str(l1.val)
            l1 = l1.next
        return str_val

    def strToNode(s1: str, s2: str) -> ListNode:
        res: ListNode = None
        out = str(int(s1) + int(s2))
        while len(out) > 0:
            cur = ListNode(out.pop())
            cur.next = res
            res = cur
        return res
    
    a = nodeToStr(reverseNode(l1))
    b = nodeToStr(reverseNode(l2))
    result = strToNode(a, b)
    return result

l1 = ListNode(1)
l2 = ListNode(2)
l3 = ListNode(3)
l1.next = l2
l2.next = l3

v1 = ListNode(1)
v2 = ListNode(2)
v3 = ListNode(3)
v1.next = v2
v2.next = v3

def printList(l1: ListNode):
    while l1:
        print(l1.val)
        l1 = l1.next
printList(addTwoNumbers(l1, v1))
```

```python
'''
using full adder
wtc : O()
wsc : O()
'''
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def addTwoNumbers(l1: ListNode, l2: ListNode) -> ListNode:
    root = head = ListNode(0)

    carry = 0
    while l1 or l2 or carry:
        sum = 0
        if l1:
            sum += l1.val
            l1 = l1.next
        if l2:
            sum += l2.val
            l2 = l2.next
        carry, val = divmod(sum + carry, 10)
        head.next = ListNode(val)
        head = head.next
    return root.next

l1 = ListNode(2)
l2 = ListNode(4)
l3 = ListNode(3)
l1.next = l2
l2.next = l3

v1 = ListNode(5)
v2 = ListNode(6)
v3 = ListNode(4)
v1.next = v2
v2.next = v3

def printList(l1: ListNode):
    while l1:                                                                                                             
        print(l1.val)
        l1 = l1.next
printList(addTwoNumbers(l1, v1))
```
